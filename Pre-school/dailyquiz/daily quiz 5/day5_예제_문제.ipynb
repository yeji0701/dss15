### 7. 달리기 시합 결과 입력 및 출력 하기 (enumerate)

- 등수를 출력할때 enumerate 를 사용해주세요.
- 결과

```
1등 한 동물은 ? 사슴
2등 한 동물은 ? 토끼
3등 한 동물은 ? 거북이

결과
1등 - 사슴
2등 - 토끼
3등 - 거북이
```
