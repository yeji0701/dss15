### 1. 역삼각형 별 출력
- 5를 입력했을때 아래와 같이 별이 출력되도록 코드를 작성하세요.
- 홀수만 입력이 가능합니다.

```
*****
 ***
  *
```

````python
number = 5
space = 0
for star in range(number, 0, -2):
    print(" " * space + "*" * star)
    space += 1
````

### 2. fizz, buzz
- 시작 숫자와 끝나는 숫자를 입력받으세요.
- while 문을 사용하세요.
- 결과 출력
    - 3의 배수는 fizz 출력
    - 5의 배수는 buzz 출력
    - 3과 5의 배수는 fizzbuzz 출력
    - 3과 5의 배수가 아니면 그냥 숫자가 출력

````python
# 시작 숫자와 끝나는 숫자를 입력받음
start = int(input("insert start number:"))
end = int(input("insert end number:"))
print()

# number 변수에 초기값을 설정
number = start

# 마지막 숫자를 초과하면 반복문 종료
while number <= end:
    
    # 3과 5의 배수인경우 "fizzbuzz" 출력
    if number % 15 == 0:
        print("fizzbuzz", end=" ")
    # 3의 배수인경우 "fizz" 출력
    elif number % 3 == 0:
        print("fizz", end=" ")
    # 5의 배수인경우 "buzz" 출력
    elif number % 5 == 0:
        print("buzz", end=" ")
    # 3과 5의 배수가 아닌경우 숫자 출력
    else:
        print(number, end=" ")
        
    # 반복문이 실행될때 마다 숫자를 1씩 증가
    number += 1
````

### 3. 가로로 구구단 출력 
- print 문에서 tab으로 띄우는 방법 : end 옵션에 "\t"를 사용
- 아래와 같이 결과가 출력되도록 반복문을 사용하여 코드를 작성하세요.

````python
# 단에 곱해지는 숫자를 출력하는 반복문
for num2 in range(0, 10):
    
    # 2단 부터 9단까지의 반복문
    for num1 in range(2, 10):
        
        # num2가 0이면 몇 단인지 출력
        if num2 == 0:
            print("{}단".format(num1), end="\t")
            
        # num2가 1이상이면 구구단의 결과를 출력
        else:
            print("{}*{}={}".format(num1, num2, num1*num2), end="\t")
    
    # 한줄 띄기
    print()
````

### 4. 피보나치 수열
- 피보나치 수열이란?
    - num1 = 0, num2 = 1의 두개의 숫자로 시작하여 num1은 num2가 되고 num2는 num1+num2가 반복적으로 실행되면서 출력되는 결과
- 피보나치 수열을 확인할 마지막 숫자를 입력하면 수열이 아래와 같이 출력되는 코드를 작성하세요.

````python
# 마지막 숫자 입력
end_number = int(input("숫자 몇까지의 피보나치 수열을 확인하겠습니까? "))

# 초기값 설정
num1, num2 = 0, 1

# num1은 num2가 되고 num2는 num1 + num2가 되도록 반복적으로 실행
while num1 < end_number:
    print(num2, end=' ')
    num1, num2 = num2, num1 + num2
````

### 5. 소수 구하기
- 소수 : 1과 자기자신으로만 나누어 떨어지는수
- 숫자를 입력받아 입력받은 숫자까지의 소수와 소수의 갯수를 출력하는 코드를 작성하세요.

````python
number = 100
count = 0

# 입력한 숫자까지 반복문으로 소수인지 확인
for num1 in range(1, number+1):
    
    # 소수인지 아닌지 판단하는 변수
    is_prime = True
    
    # 소수를 확인하기 위해 확인하려는 숫자보다 작은 수를 나누어봄
    for num2 in range(2, num1):
        
        # 소수 조건 확인
        if num1 % num2 == 0:
            
            # 소수가 아님
            is_prime = False
            
            # 반복문 중단
            break
            
    # is_prime이 True이면 num1이 소수
    if is_prime:
        
        # 소수 카운트 증가
        count += 1
        
        # 소수 출력
        print(num1, end=" ")
        
# 소수의 갯수 출력
print("\n소수의 갯수 : {}개".format(count))
````
