```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2 * np.sin(x + 0.5)
y3 = 3 * np.sin(x + 1.0)
y4 = 4 * np.sin(x + 1.5)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
sns.set_style("white")
plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
sns.despine(); plt.show()
```

```python
sns.set_style("dark")
plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
sns.despine(offset=10)
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=tips['total_bill'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='Set2');
```

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x='day', y='total_bill', data=tips, color='.5');
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='.75');
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips, height=6);
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips, height=10);
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips);
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights = flights.pivot('month', 'year', 'passengers')
flights.head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt='d');
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt='d', cmap='YlGnBu');
```
