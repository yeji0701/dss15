```python
import requests
import json
import datetime as dt
from datetime import timedelta, datetime
import pandas as pd
```

```python
result = {}

for i in range(261):

    key = ''
    targetDt = datetime(2020, 11, 4) - timedelta(weeks=i)
    targetDt = targetDt.strftime('%Y%m%d')

    url = f'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchWeeklyBoxOfficeList.json?key={key}&targetDt={targetDt}&multiMovieYn=N&repNationCd=K'
    
    response = requests.get(url)
    data = response.json()['boxOfficeResult']['weeklyBoxOfficeList']
    
    for movie in data:
        code = movie.get('movieCd')
        if code not in result: 
            result[code] = {
                'Code' : movie.get('movieCd'),
                'Movie': movie.get('movieNm'),
                'Audience': movie.get('audiAcc'),
                'Open' : movie.get('openDt'),
                'Screen' : movie.get('scrnCnt')
            }
```

```python
df = pd.DataFrame(result)
df
```

```python
codes = df.loc["Code"]

code_list = []

for code in codes:
    code_list.append(code)
```

```python
result2 = {}

for movieCd in code_list:
    
    key = ''
    code = movieCd

    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key={}&movieCd={}'.format(key, code)
    
    response = requests.get(url)
    data = response.json()['movieInfoResult']['movieInfo']

    for movie_info in data:
        
        result2[movieCd] = {
            'Code': data.get('movieCd'),
            'watchGradeNm': data.get('audits')[0].get('watchGradeNm') if data.get('audits') else None, 
            'genres': data.get('genres')[0].get('genreNm') if data.get('genres') else None,
        }
        
result2
```

```python
df2 = pd.DataFrame(result2)
df2
```

```python
movie_df = pd.concat([df, df2])
movie_df
```

```python
movie_df = movie_df.transpose()
movie_df
```

```python
movie_df.drop_duplicates(inplace=True)
movie_df
```

```python
movie_df.drop(columns='Code', inplace=True)
movie_df
```

```python
movie_df.to_csv('./kobis_movie.csv', sep=',', encoding='utf-8')
```
