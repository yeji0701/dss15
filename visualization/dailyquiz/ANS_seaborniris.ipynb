```python
# enviroment setting 

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
# load dataset
iris_df = sns.load_dataset('iris')
```

```python
# data -- iris
iris_df.head()
```

# Quiz 1. make pivot table 
- index ='species'
- aggfunc = np.mean

```python
iris_df.pivot_table(iris_df.columns,'species',aggfunc = np.mean)
```

# Quiz 2. make boxplot
        - x_values = sepal_length , y_values  = petal_width
        - hue = Species    

```python
plt.figure(figsize = (14,8))
# code 
sns.boxplot(x = iris_df['sepal_length'], y = iris_df['petal_width'],hue = iris_df["species"],palette = 'Set3')
```

#  Quiz 3. make lmplot
    - x_values = sepal_length , y_values  = petal_width
    - hue = Species    

```python
plt.figure(figsize = (12,8))

sns.lmplot(x = 'sepal_length' , y = 'petal_width',data = iris_df ,hue = 'species')
```

#  Quiz 4. make pairplot
    - x_values = sepal_length','sepal_width, y_values  = petal_width
    - hue = species    

```python
plt.figure(figsize = (14,8))


sns.pairplot(iris_df,x_vars = ['sepal_length','sepal_width'] , y_vars = ['petal_width'], hue = 'species')
```

#  Quiz 5. make plot 
    - drop 'species' column
    - use np.exp(data)
    - figsize = (16,10)

```python
np.exp(iris_df.drop('species',1),).plot(figsize = (16,10))
```
