# Quiz 2. make boxplot
- iris_df를 활용해서 boxplot을 그려주세요!
- x_values = sepal_length , y_values  = petal_width 
- 종으로 분류 해주세요!   

```python
sns.set(style='darkgrid')
plt.figure(figsize = (14,8))
sns.boxplot(x='sepal_length', y='petal_width', hue='species', data=iris_df)
plt.legend(title='species', loc='best')
plt.show()
```

#  Quiz 3. make lmplot
- iris_df 를 활용해서 lmplot을 그래주세요!
- x_values = sepal_length , y_values  = petal_width
- lmplot은 종에 따라 구별되게 해주세요!   

```python
# sns.set_style('dark')
sns.set(style='darkgrid')
plt.figure(figsize = (12,8))
sns.lmplot(x='sepal_length', y='petal_length', hue='species', data=iris_df)
plt.show()
```

```python
plt.figure(figsize = (12,8))

# code
```

#  Quiz 4. make pairplot
- iris_df를 활용하여서 pairplot을 그려주세요!
- x_values = sepal_length','sepal_width, y_values  = petal_width
- 종으로 분류 해주세요!    

```python
sns.set(style='darkgrid')
sns.pairplot(iris_df, x_vars=['sepal_length', 'sepal_width'],
             y_vars=['petal_width'], hue='species')
plt.show()
```

```python
plt.figure(figsize = (14,8))

# code
```
