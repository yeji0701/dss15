```python
import pandas as pd
import numpy as np
```

```python
Data = pd.DataFrame({
    'age' : np.arange(30,55),'sales': np.repeat([60,40,30,0,20],5)
                     })
```

```python
Data2 = pd.DataFrame({
    'age' : np.arange(30,55),'freq': np.repeat([10,5,3,2,4],5)
                     })
```

```python
# 문제 1 정답
Data_mean = Data['sales'].mean()
Data["mean"] = Data['sales'] - Data_mean
Data

```

```python
# 문제 2 정답 
pd.merge(Data,Data2)[["age","sales","freq"]]
```

```python
# 문제3 정답
Data3 = pd.merge(Data,Data2)

Data3['con'] = Data3['sales'] > 30  
Data3
```
