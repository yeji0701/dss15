```python
def get_stock_data(code, page_size=20, page=1):
    url = 'https://m.stock.naver.com/api/json/sise/dailySiseIndexListJson.nhn?code={}&pageSize={}&page={}'.format(code, page_size, page)
    response = requests.get(url)
    response.json()
    data = response.json()['result']['siseList']
    
    return pd.DataFrame(data)
```

```python
from bs4 import BeautifulSoup
```

```python
query = '삼성전자'
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query={}'.format(query)
```

```python
response = requests.get(url)
```

```python
dom = BeautifulSoup(response.content, 'html.parser')
```

```python
elements = dom.select('.lst_related_srch > li')
len(elements), elements[0]
```

```python
keywords = []
for element in elements:
    keywords.append(element.text.strip())

keywords
```

```python
keywords = [element.text.strip() for element in elements]
keywords
```

```python
def naver_relate_keywords(query):
    url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query={}'.format(query)
    response = requests.get(url)
    dom = BeautifulSoup(response.content, 'html.parser')
    elements = dom.select('.lst_related_srch > li')
    
    return [element.text.strip() for element in elements]
```

```python
query = '파이썬'
result = naver_relate_keywords(query)
result
```

```python
url = 'https://news.daum.net/'
response = requests.get(url)
dom = BeautifulSoup(response.content, 'html.parser')
```

```python
elements = dom.select('.list_headline > li')
len(elements), elements[0]
```

```python
datas = []
for element in elements:
    datas.append({
        "title" : element.select_one('.tit_g').text,
        "link" : element.select_one('a').get('href')
    })
datas
```

```python
articles_df = pd.DataFrame(datas)
articles_df.tail()
```

```python
link = articles_df['link'][0]
```

```python
response = requests.get(link)
dom = BeautifulSoup(response.content, 'html.parser')
```

```python
content = dom.select_one('#harmonyContainer').text.strip().replace('\n', '')
```

```python
def get_content(link):
    response = requests.get(link)
    dom = BeautifulSoup(response.content, 'html.parser')
    return dom.select_one('#harmonyContainer').text.strip().replace('\n', '')
```

```python
articles_df['contents'] = articles_df['link'].apply(get_content)
```
